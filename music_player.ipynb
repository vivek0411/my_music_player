```python
from tkinter import *
from os import system
from tkinter import filedialog
import pygame
from pygame import mixer
mixer.init()


# main page
root=Tk()
root.title("Music Player")
root.iconbitmap(r'c:\users\vivek singh\downloads\music.ico')
root.minsize(500,500)
root.config(bg="lightskyblue")

audioTrack=StringVar()

def musicurl():
    dd=filedialog.askopenfilename()
    audioTrack.set(dd)
    
def playmusic():
    ad=audioTrack.get()
    mixer.music.load(ad)
    mixer.music.play()
   

def pausemusic():
    mixer.music.pause()

def resume():
    mixer.music.unpause()
    
    
    
def volup():
    vol= mixer.music.get_volume()
    mixer.music.set_volume(vol + 0.1)

def voldown():
    vol= mixer.music.get_volume()
    mixer.music.set_volume(vol - 0.1)

def stopmusic():
    mixer.music.stop()
    

# label,entry,button

l1=Label(root,text="Select Audio Track : ",font="Arial 15 bold")
l1.grid(row=0,column=0,padx=20,pady=20)


e1=Entry(root,font="Arial 16 bold",textvariable=audioTrack)
e1.grid(row=0,column=1,padx=20,pady=20)

b1=Button(root,text="Search",font="Arial 13 bold",bg="deeppink",activebackground="purple",width=20,command=musicurl)
b1.grid(row=0,column=3,padx=20,pady=20)

b2=Button(root,text="Play",font="Arial 13 bold",bg="green2",activebackground="purple",width=20,command=playmusic)
b2.grid(row=1,column=0,padx=20,pady=20)



###### some extra buttons


root.b3=Button(root,text="Pause",font="Arial 13 bold",bg="deeppink",activebackground="purple",width=20,command=pausemusic)
root.b3.grid(row=1,column=1,padx=20,pady=20)

root.b6=Button(root,text="Resume",font="Arial 13 bold",bg="deeppink",activebackground="purple",width=20,command=resume)
root.b6.grid(row=2,column=1,padx=20,pady=20)



b4=Button(root,text="Volume Up",font="Arial 13 bold",bg="deeppink",activebackground="purple",width=20,command=volup)
b4.grid(row=1,column=3,padx=20,pady=20)

b5=Button(root,text="Stop",font="Arial 13 bold",bg="deeppink",activebackground="purple",width=20,command=stopmusic)
b5.grid(row=2,column=0,padx=20,pady=20)

b1=Button(root,text="Volume Down",font="Arial 13 bold",bg="deeppink",activebackground="purple",width=20,command=voldown)
b1.grid(row=2,column=3,padx=20,pady=20)





root.mainloop()
```
